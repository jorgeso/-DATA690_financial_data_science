```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

tickers = ['^GSPC', 'NDAQ', 'BTC-USD', 'ETH-USD']
start = datetime.datetime(2015,1,1)
end = datetime.date.today()
stocks = yf.download(tickers, start=start,end = end, interval='1d')['Close']
stocks
```

```python
stocks = stocks.rename(columns={'^GSPC': 'S&P 500', 'NDAQ': 'NASDAQ', 'BTC-USD': 'Bitcoin', 'ETH-USD': 'Ethereum'})
stocks.plot(figsize=(20,10), logy=True)
```

```python
import time
from IPython import display
import pylab as pl
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from dateutil import tz

%matplotlib inline
plt.figure(figsize=(20,10))
to_zone = tz.gettz('America/New_York')
while True:
    tickers = ['BTC-USD']
    start = datetime.datetime(2020, 10, 1)
    end = datetime.datetime.now()
    btc = yf.download(tickers, start=start,end = end, interval='1h')['Close']
    now = btc.index[-1].astimezone(to_zone)
    plt.title(f'{now}: ${round(btc.values[-1], 2):,}')
    pl.plot(btc)
    display.clear_output(wait=True)
    display.display(pl.gcf())
    time.sleep(60)
```
